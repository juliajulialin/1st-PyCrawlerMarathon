### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find('span').text, d.find('em').text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
data = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    cag = d.find('em').text
    date = d.find('span').text
    title = d.find_all('a')[-1].text
    data.setdefault(cag, [])
    data[cag].append({
        'date': date,
        'title': title
    })
    
print(data)
```

### ③ 哪一個類別的文章最多

```python
print('文章最多的類別是：')
max_art=0
cag
for d in data:
    if max_art < len(data[d]):
        max_art = len(data[d])
        cag = d

print(cag, max_art)
```
